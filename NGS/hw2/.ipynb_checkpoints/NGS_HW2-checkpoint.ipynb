#### 1.Покрытие генома

```python
# считает список, где каждый i-й элемент соответсвует числу ридов, покрывающих позицию i
def get_coverage(file, length_ref):
    bamfile = pysam.AlignmentFile(file, 'rb')
    positions = [0 for i in range(length_ref)]

    for read in bamfile.fetch():
        for pos in read.get_reference_positions():
            positions[pos] += 1

    bamfile.close()
    pos_average = []
    curr = 0
    step = 3000
    ave_coverage = 0
    per_coverage = 0
    for i in range(len(positions)):
        if positions[i] != 0:
            per_coverage += 1
        ave_coverage +=  positions[i]
        if i % step == 0 and i != 0:
            pos_average.append(curr / step)
            curr = 0
        curr +=  positions[i]
    pos_average.append(curr / (len(positions) % step))
    
    ave_coverage /= len(positions)
    per_coverage /= len(positions)
    
    print('average coverage of genome: ', ave_coverage)
    print('proportion of the covered genome: ', per_coverage)
    
    plt.plot([i for i in range(len(pos_average))], pos_average)
    plt.gca().set(xlabel='average positions', ylabel='reads')
    plt.show()
    return positions

def ref_len(path):
    print(path)
    f=open('path', 'r')
    s = ''
    for line in f:
        if line[0] != '>':
            s += line

    f.close()
    return len(s)
```

#### 2. Распределение расстояния вставки.

```python
# считает расстояние вставки для каждого рида
def dist_isize(file):
    bamfile = pysam.AlignmentFile(file, 'rb')    
    isizes = []
    for read in bamfile.fetch():
        if read.tlen != 0:
            isizes.append(abs(read.tlen))
    bamfile.close()
    sns.distplot(isizes)
    return isizes

# считает интервал минимальной ширины, в который попадает 95% ридов
def get_interval(dist):
    x = np.mean(dist)
    sigma = np.std(dist)
    n = len(dist)
    q = scipy.stats.norm.ppf(0.95)
    
    return x - sigma / n * q, x + sigma / n * q

# по распределению считает некоротые характеристики
def dist_isize_characteristic(dist):
    print('average insert size: ', np.mean(dist))
    print('std: ', np.std(dist))
    a, b = get_interval(dist)
    print('confident interval: (', a, '; ', b,')')
```

#### 3. Частоты различных замен


```python
# счиатет матрицу частот замен 
def get_freq_replacement(file):
    nucls = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
    matrix = np.zeros((4, 4))
    bamfile = pysam.AlignmentFile(file, 'rb')
    for read in bamfile.fetch():
        if read.is_unmapped != True:
            bases = read.get_aligned_pairs(matches_only=True, with_seq=True)
            for base in bases:
                if read.seq[base[0]] != base[2] and read.seq[base[0]] != 'N':
                     matrix[nucls[read.seq[base[0]]]][nucls[base[2].upper()]] += 1
    print(matrix)
    bamfile.close()

print('100x:')    
get_freq_replacement('./e.coli/100x_alignment.sam')
    
    
print('\necoli:')    
get_freq_replacement('./e.coli/ecoli_alignment_sorted.bam')
```

#### Средний процент ошибок в ридах:

```python
def get_errors_percent(file):
    err_percent = []
    bamfile = pysam.AlignmentFile(file, 'rb')
    for read in bamfile.fetch():
        err_percent.append(read.get_tag('NM') / read. query_alignment_length) # NM - edit distance, is it fine???
    bamfile.close()
    
    print('Average percent of errors: ', np.mean(err_percent) * 100, "%")
get_errors_percent('./data/alignment_sorted.bam')
```

#### Статистика инделов:

```python
def get_indels_stat(file):
    insertions = []
    deletions = []
    bamfile = pysam.AlignmentFile(file, 'rb')
    
    for read in bamfile.fetch():
        for tup in read.cigartuples:
            if tup[0] == 1: # insertions
                insertions.append(tup[1])
            elif tup[0] == 2: # deletions
                deletions.append(tup[1])
    bamfile.close()
    
    print('average number of insertions in read: ', np.mean(insertions))
    print('std of insertions in read: ', np.std(insertions))
    a, b = get_interval(insertions)
    print('confident interval, 95%: (', a, ';', b, ')')
    
    print('\naverage number of deletions in read: ', np.mean(deletions))
    print('std of deletions in read: ', np.std(deletions))
    c, d = get_interval(deletions)
    print('confident interval, 95%: (', c, ';', d, ')')
    
get_indels_stat('./data/alignment_sorted.bam')
```
