## BRCA1

Был выбран ген BRCA1, немного информации о выбранных видах:

Erinaceus amurensis - Амурский ёж (ареал: cеверный Китай, Корейский полуостров, Восточная Россия)

Erinaceus europaeus - Обыкновенный ёж (ареал: Западная и Центральная Европа, Британские острова, северо-запад Европейской части России, Западная Сибирь, Казахстан)

Erinaceus concolor - Восточноевропейский ёж (ареал: распрастранен от Центральной Европы до Западной Сибири)

Paraechinus aethiopicus - Эфиопский ёж (ареал: Северная Африка, Сирия, Ирак, Иран)

Рaraechinus hypomelas - Длинноиглый ёж (ареал: Иран, Пакистан, Афганистан, Туркмения)

Atelerix frontalis - Южноафриканский ёж (ареал: Ангола, Ботсвана, Лесото, Намибия, Зимбабве и ЮАР)

Atelerix albiventris - Белобрюхий ёж (ареал:  юг Сахары, Мавритания, Сенегал, Нигерия, Судан, Эфиопия)

Atelerix algirus - Алжирский ёж (ареал: Алжир, Франция, Ливия, Марокко, Испания и Тунис)

Mesechinus dauuricus - Даурский ёж (ареал: юго-восток Забайкальского края, северо-восточная Монголия и Северный Китай)

Hemiechinus auritus - Ушастый ёж (ареал: Израиль, Ливию, Египет, частично Индия, Казахстан, Китай и Монголия)

Hylomys suillus - Малая гимнура (ареал:  Бруней, Вьетнам, Индонезия, Камбоджа, Китай, Лаос, Малайзия, Мьянма, Таиланд)

Podogymnura truei - эндемики, ареал: острава Минадамо, Филиппины. Являются одними из старейших известных плацентарных млекопитающих, которые живы.

Neotetracus sinensis - ареал: Китай, Мьянма, Вьетнам
