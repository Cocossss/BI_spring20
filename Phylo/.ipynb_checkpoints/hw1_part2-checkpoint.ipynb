```python
import requests
import random
from Bio import Phylo
from io import StringIO
import matplotlib.pyplot as plt
from matplotlib import pylab
from ete3 import Tree, TreeStyle, NodeStyle, faces, AttrFace, CircleFace
```

## Part 1

```python
response = requests.get('https://www.jasondavies.com/tree-of-life/life.txt')
tree = response.text # get newick format
tree = next(Phylo.parse(StringIO(tree), 'newick'))
Phylo.draw_ascii(tree)
```

```python
plt.rcParams.update({'font.size': 10, 'lines.linewidth': 1})
#print(plt.rcParams)
plt.rcParams['figure.figsize'] = 30, 30 # size of picture
Phylo.draw(tree, do_show=False)
plt.axis('off') # remove coordinates
```

```python
Phylo.write(tree, 'example.xml', 'phyloxml')
```

## Part 2

```python
t = Tree(response.text, format=1)
t.show()
```

```python
leaves = t.get_leaves()
inds = [] # list for indices of 42 random leaves
for i in range(42):
    inds.append(random.randint(0, len(leaves)-1))
    
to_prune = [] # list for 42 random leaves
for el in inds:
    to_prune.append(leaves[el])
t.prune(to_prune)
tmp = t.render('prunned.png')
```

```python
nst = NodeStyle()
nst['fgcolor'] = 'green' # all nodes are green
nst['hz_line_color']  = 'green' # all edges are green
nodes = t.traverse() # list of all nodes
for el in nodes:
    el.set_style(nst)

tmp = t.render('prunned_and_green.png')
```
